## Quest√£o 3) 

Considerando os dados presentes no arquivo class01.csv, treine o algoritmo Naive Bayes Gaussiano utilizando a metodologia de valida√ß√£o cruzada holdout (utilize para treino as 350 primeiras linhas e para valida√ß√£o as demais). Qual o valor da acur√°cia para a base de treino? Qual o valor da acur√°cia para a base de valida√ß√£o? Fa√ßa o mesmo treinamento com a metodologia Leave-One-Out. Qual o valor da acur√°cia m√©dia para a base de treino? Qual o valor da acur√°cia m√©dia para a base de valida√ß√£o?

```python
class01 = pd.read_csv('class01.csv')
```

### Hold Out

```python
X = class01.drop(columns=['target'])
y = class01['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.65, shuffle=False)
```

```python
X_train
```

```python
clf = GaussianNB()
clf.fit(X_train, y_train)
print('Acur√°cia - treino:', + clf.score(X_train, y_train))
print('Acur√°cia - valida√ß√£o:', + clf.score(X_test, y_test))
```

### Leave One Out

```python
X = class01.drop(columns=['target'])
y = class01['target']
cross = cross_validate(GaussianNB(), X, y, cv=LeaveOneOut(), return_train_score=True)

print('Acur√°cia m√©ida - treino:', + cross['train_score'].mean())
print('Acur√°cia m√©ida - teste:', + cross['test_score'].mean())
```

Considerando os dados presentes no arquivo classificacao_Q3.csv, treine o algoritmo Naive Bayes utilizando a metodologia de valida√ß√£o cruzada holdout (utilize para treino as 500 primeiras linhas e para valida√ß√£o as demais) Qual o valor da acur√°cia a base de treino? Qual o valor da acur√°cia m√©dia para as bases de valida√ß√£o?

```python
classe_Q3 = pd.read_csv('classificacao_Q3.csv')
classe_Q3
```

```python
X = classe_Q3.drop(columns=['Target'])
y = classe_Q3['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=500, shuffle=False)
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder()
clf = CategoricalNB()
clf.fit(encoder.fit_transform(X_train), y_train)
print('Acur√°cia - treino:', + clf.score(encoder.fit_transform(X_train), y_train))
print('Acur√°cia - valida√ß√£o:', + clf.score(encoder.transform(X_test), y_test))
```
