#### Question 1: Find the questions and answers of last month.

```python
import requests
from datetime import datetime
```

```python
# your code here
questions = site.fetch('questions', fromdate = datetime(2020,2,1), todate= datetime(2020,2,29))
questions
```

```python
answers = site.fetch('answers', fromdate = datetime(2020,2,1), todate= datetime(2020,2,29))
answers
```

#### Question 2: Find the most voted question today with at least a score of 5 and tagged with 'python'. 

```python
# your code here
questions_today = site.fetch('questions', fromdate = datetime(2020,3,22))
questions_today
```

```python
voted =[]
pattern = 'python'
for i in range(len(questions_today['items'])):
    if pattern in questions_today['items'][i]['tags']:
        if questions_today['items'][i]['owner']['reputation'] > 4:
            print(questions_today['items'][i]['title'],questions_today['items'][i]['owner']['reputation'])
```

#### Question 3: Find the answers with id 6784 and 6473.

```python
# your code here
pattern1 = '6784'
pattern2 = '6473'
for i in range(len(answers['items'])):
    if pattern1 in str(answers['items'][i]['answer_id']):
        print('6784', answers['items'][i]['owner']['link'])
    elif pattern2 in str(answers['items'][i]['answer_id']):
        print('6473', answers['items'][i]['owner']['link'])
    else:
        'Nada'
```
