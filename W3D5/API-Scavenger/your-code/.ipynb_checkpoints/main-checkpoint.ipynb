## Challenge 1: Fork Languages

You will find out how many programming languages are used among all the forks created from the main lab repo of your bootcamp.

`````python
# import libraries here
import requests
`````

Assuming the main lab repo is ironhack-datalabs/mad-oct-2018, you will:

#### 1. Obtain the full list of forks created from the main lab repo via Github API.

To list forks, we can use the GET method. As explained in the GitHub API documentation, we need to make the request to: GET /repos/:owner/:repo/forks.

`````python
# your code here
from requests.auth import HTTPBasicAuth
response = requests.get("https://api.github.com/repos/ironhack-datalabs/mad-oct-2018/forks",
                        auth=HTTPBasicAuth('user', 'token'))
response
`````

#### 2. Loop the JSON response to find out the language attribute of each fork. Use an array to store the language attributes of each fork.
Hint: Each language should appear only once in your array.
Print the language array. It should be something like: ["Python", "Jupyter Notebook", "HTML"]

`````python
# your code here
results = response.json()
import numpy as np
languages = []
for i in range(len(results)):
    if results[i]['language'] != None:
        languages.append(results[i]['language'])
np.array(set(languages))
`````

#### 1. Obtain all the commits made in January 2019 via API, which is a JSON array that contains multiple commit objects.

`````python
# your code here
response = requests.get("https://api.github.com/repos/ironhack-datalabs/mad-oct-2018/commits",
                        auth=HTTPBasicAuth('user', 'token'))
results = response.json()
date = []
pattern = '2019-01'
for i in range(len(results)):
    if pattern in results[i]['commit']['author']['date']:
        date.append(results[i]['commit']['author']['date'])
jan = np.array(date)
jan
`````

## Challenge 3: Hidden Cold Joke

Using Python, call Github API to find out the cold joke contained in the 24 secret files in the following repo:

https://github.com/ironhack-datalabs/scavenger

The filenames of the secret files contain .scavengerhunt and they are scattered in different directories of this repo. The secret files are named from .0001.scavengerhunt to .0024.scavengerhunt. They are scattered randomly throughout this repo. You need to search for these files by calling the Github API, not searching the local files on your computer.

#### 1. Find the secret files.

`````python
# your code here
response = requests.get("https://api.github.com/repos/ironhack-datalabs/scavenger/contents/", 
                        auth=HTTPBasicAuth('user', 'token'))
results = response.json()
file = []
for t in range(len(results)):
    if results[t]['name'] != '.gitignore':
        file.append(results[t]['name'])
    
secret =[]
for i in range(len(file)):
    secret.append(requests.get("https://api.github.com/repos/ironhack-datalabs/scavenger/contents/"+file[i], 
                               auth=HTTPBasicAuth('user', 'token')))
    secret[i] = secret[i].json()
secret
`````

`````python
files = []
pattern = 'scavengerhunt'
for i in range(len(secret)):
    for j in range(len(secret[i])):
        if pattern in secret[i][j]['name']:
            files.append(secret[i][j]['name'])
files
`````

#### 2.  Sort the filenames ascendingly.

`````python
# your code here
files.sort(reverse=False)
`````

`````python
files
`````

#### 3. Read the content of each secret files into an array of strings.
Since the response is encoded, you will need to send the following information in the header of your request:
````python
headers = {'Accept': 'application/vnd.github.v3.raw'}
````

`````python
files_str = []
pattern = 'scavengerhunt'
for i in range(len(secret)):
    for j in range(len(secret[i])):
        if pattern in secret[i][j]['path']:
            files_str.append(secret[i][j]['path'])
files_str.sort(key=lambda x: x.split('/.')[1].split('.')[0])
files_str
`````

`````python
# your code here
headers = {'Accept': 'application/vnd.github.v3.raw'}
frase = []
for i in range(len(files_str)):
    response = requests.get("https://api.github.com/repos/ironhack-datalabs/scavenger/contents/"+files_str[i], 
                               auth=HTTPBasicAuth('user', 'token'), headers=headers)
    frase.append(response.text)
frase
`````

#### 4. Concatenate the strings in the array separating each two with a whitespace.

`````python
# your code here
joke = ' '.join(frase).replace('\n', '')
`````

#### 5. Print out the joke.

`````python
# your code here
print(joke)
`````
