#### Select all the customers from Spain.
**Hint**: Remember that you are not asked to find orders from Spain but customers. A customer might have more than one order associated. 

```python
# your code here
orders.loc[orders['Country'] == 'Spain']['CustomerID'].unique()
```

#### How many customers do we have in Spain?

```python
# your code here
len(orders.loc[orders['Country'] == 'Spain']['CustomerID'].unique())
```

#### Select all the customers who have bought more than 50 items.
**Hint**: Remember that you are not asked to find orders with more than 50 items but customers who bought more than 50 items. A customer with two orders of 30 items each should appear in the selection.

```python
# your code here
quantity_50 = orders.groupby(['CustomerID'])[['Quantity']].sum()
mask = orders.groupby(['CustomerID'])[['Quantity']].sum() > 50
quantity_50[mask].dropna()
```

#### Select orders from Spain that include more than 50 items.

```python
# your code here
quantity_50_sp = orders.loc[orders['Country'] == 'Spain'].groupby(['CustomerID'])[['Quantity']].sum()
mask = orders.loc[orders['Country'] == 'Spain'].groupby(['CustomerID'])[['Quantity']].sum() > 50
quantity_50_sp[mask]
```

#### Select all free orders.

```python
# your code here
orders.loc[orders['amount_spent'] == 0.00]
```

#### Select all orders whose description starts with `lunch bag`.
**Hint**: use string functions.

```python
# your code here
orders.loc[orders['Description'].str.startswith('lunch bag')]
```

#### Select all `lunch bag` orders made in 2011.

```python
# your code here
orders.loc[orders['Description'].str.startswith('lunch bag')].query('year == 2011')
```

#### Show the frequency distribution of the amount spent in Spain.

```python
# your code here
orders.loc[orders['Country'] == 'Spain',:]['amount_spent'].value_counts()
```

```python
plt.hist(orders.loc[orders['Country'] == 'Spain',:]['amount_spent'].value_counts(),bins=100, color = 'magenta')
plt.xlabel('Amount Spent')
plt.ylabel('Frequency')
plt.show()
```

#### Select all orders made in the month of August.

```python
# your code here
orders.loc[orders['month'] == 8]
```

#### Find the number of orders made by each country in the month of August.
**Hint**: Use value_counts().

```python
# your code here
orders.loc[orders['month'] == 8].groupby(['Country'])[['amount_spent']].count()
```

#### What's the  average amount of money spent by country?

```python
# your code here
orders.loc[orders['month'] == 8].groupby(['Country'])[['amount_spent']].mean()
```

#### What's the most expensive item?

```python
# your code here
orders.groupby(['Description'])[['UnitPrice']].max().sort_values(by='UnitPrice', ascending=False).head(1)
```

#### What is the average amount spent per year?

```python
# your code here
orders.groupby(['year'])[['amount_spent']].mean()
```
