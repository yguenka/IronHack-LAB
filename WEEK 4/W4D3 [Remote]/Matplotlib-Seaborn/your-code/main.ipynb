```python
# Method 1
plt.plot(x, z, label = 'LINEAR', color = 'purple', linewidth = 3) 
plt.plot(x, y, label= 'QUADRATICS', color = '#ff008c', linewidth = 3) #color picker

plt.legend()
plt.title('My first plot !')
plt.show()
```

#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, z, label = 'LINEAR', color = 'purple', linewidth = 3) 
ax2.plot(x, y, label= 'QUADRATICS', color = '#ff008c', linewidth = 3)
plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2, figsize =(5,5))
ax1.plot(x, y, color = 'purple', linewidth = 5)
ax1.set_title('LINEAR PLOT')

ax2.plot(x, z, color = '#ff008c', linewidth = 5)
ax2.set_title('QUADRATIC PLOT')
plt.show()
```

#### Now you will have to plot both $y=x^2$ and $y=e^x$ in the same plot. Create two subplots and use a normal scale in the first one and a logarithmic scale in the second one.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# your code here
import math
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, x**2, color = 'purple', linewidth = 5)

ax2.plot(x, math.e**x, color = '#ff008c', linewidth = 5)
ax2.set_xscale("log")

plt.show()
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# your code here
fig, [ax1, ax2] = plt.subplots(nrows=1, ncols=2)
ax1.plot(x, x**2, color = 'purple', linewidth = 5)
ax1.set_title('Escala normal')

ax2.plot(x, math.e**x, color = '#ff008c', linewidth = 5)
ax2.set_xscale("log")
ax2.set_yscale("log")
ax2.set_title('Escala logartmica')
plt.show()

```
