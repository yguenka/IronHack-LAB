```python
#Sub Problem 2:
orders_new.quantile(np.arange(0,1.1, 0.25))
```

```python
#Sub Problem 3:
q75 = orders_new.quantile(0.75)
orders_new['label'] = np.where(orders_new > q75, 'Yes', 'No')
orders_new.replace('CustomerID', "1")
```
